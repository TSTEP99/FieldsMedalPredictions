```python
#Import neccessary libraries
import numpy as np;
import pandas as pd;
from datetime import date,timedelta;
from math import isnan
from statistics import mode,mean;
from copy import deepcopy
```

```python
#read in mathmaticians.cdv
df=pd.read_csv("mathematicians.csv");
#Gets rid of instance of attribute
df=df.drop("instance of",axis=1)
df=df.drop("approx. date of birth",axis=1);
df=df.drop("approx. date of death",axis=1);
```

```python
#makes a copy of a pandas dataframe for the day,month and year of death based on 
birth_day=deepcopy(df["day of birth"]);
birth_month=deepcopy(df["month of birth"]);
birth_year=deepcopy(df["year of birth"]);
#initialize empty numpy array 
birth_date=np.empty(len(birth_day),dtype=date);
```

```python
print(birth_day[1],birth_month[1],birth_year[1])
#Loops through data to transform dates before 0 AD into negative integers
for i in range(len(birth_year)):
	if not isinstance(birth_year[i],str):
		continue;
	if birth_year[i][-1]=="s":
		birth_year[i]="1";
```

```python
#Function to map the month to an integer
def string2int(month):
    #print(month);
    months={"January":1,"February":2,"March":3,"April":4,"May":5,"June":6,"July":7,"August":8,"September":9,"October":10,"November":11,"December":12}
    
    return months[month];
#print(birth_day[1],birth_month[1],birth_year[1])
for i in range(len(birth_date)):
    birth_month[i]=string2int(birth_month[i]);
    birth_date[i]=date(int(birth_year[i]),birth_month[i],int(birth_day[i]));
    print(birth_date[i]);
```

```python
for i in range(len(death_year)):
	if not isinstance(death_year[i],str):
		continue;
	if death_year[i][-1]=="s":
		death_year[i]="1";
```

```python
impute_day= mode(death_day.dropna());
impute_month= mode(death_month.dropna());
impute_year= mode(death_year.dropna());



for i in range(len(death_date)):
    if isnan(death_day[i]) and not isinstance(death_month[i],str) and not isinstance(death_year[i],str):
        temp_date=birth_date[i]+timedelta(days=122*365+164);
        death_day[i]=temp_date.day;
        death_month[i]=temp_date.month;
        death_year[i]=temp_date.year
        continue;
    if isnan(death_day[i]):
        death_day[i]=impute_day
    if not isinstance(death_month[i],str):
        death_month[i]=string2int(impute_month);
    else:
        death_month[i]=string2int(death_month[i]);
    if not isinstance(death_year[i],str):
        death_year[i]=impute_year;
for i in range(len(death_date)):
    death_date[i]=date(int(death_year[i]),death_month[i],int(death_day[i]));
```

```python
def convert_array(word):
	word=word.replace("["," ");
	word=word.replace("]"," ");
	word=word.replace("'","");
	array=word.split(",");
	for i in range(len(array)):
		array[i]=array[i].strip();
	return array;
        
	return array;
def counter(array):
	array=array.dropna();
	array=np.array(array);
	count={};
	for i in range(len(array)):
		for death in convert_array(array[i]):
			count[death]=1;
	#print();
	#for death in count:
		#print(death,count[death]);
	return count
```

```python
def make_dummy(gender,names):
    gender=deepcopy(gender);

    gender_map=counter(gender);

    columns=len(gender_map);
    rows=len(gender);
    
    col_names=list(gender_map.keys());
    
    for i in range(len(col_names)):
        if col_names[i] in names:
            col_names[i]= col_names[i]+" ("+gender.name+")";
            

    dummy= pd.DataFrame(np.zeros((rows,columns),dtype=int),columns=list(gender_map.keys()));

    for i in range(len(gender)):
        array=convert_array(gender[i]);
        for item in array:
            dummy[item][i]=1;
    dummy.columns=np.array(col_names,dtype=str);
    return dummy;


    
    
    



    
```
